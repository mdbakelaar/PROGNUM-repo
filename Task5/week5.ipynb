```python
import numpy as np
from matplotlib import pyplot as plt

#importing the file
file = "ngc6946"
data = np.loadtxt(file, comments = '#')

#making an array of the file and shaping into the correct dimensions and taking a singular plane
filearray = np.array(data).reshape(40, 50, 70)
plane = filearray[0, :, :]

#getting the mean and stdev so I can use those for the contours later
meanP = np.mean(plane)
stdevP = np.std(plane)
levels = [meanP,meanP + stdevP , meanP + 2*stdevP]

#plottin the image with the colorbar and the contours
plt.imshow(plane, origin = 'lower')
plt.colorbar()
plt.contour(plane, levels = levels , colors="white", linewidths = 1)
plt.show()
```

```python
import sys
print(f"Python version: {sys.version}")

import numpy as np
print(f"NumPy version: {np.__version__}")

from scipy.constants import c

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      
    cms = c
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0) / self.caH
        return z
    def velocity(self):
        ckms = self.cms *1e-3
        v = self.redshift() * ckms
        return v


g1 = Galaxy("NGC1832", 3994.7)
print(f"The redshift of the galaxy is: {g1.redshift()}")
print(f"The correspondin velocity is: {g1.velocity()}")
```

```python
from matplotlib.patches import Ellipse
from scipy.constants import pi
import math


class AstroEllipse(Ellipse):
   def __init__(self, xc, yc, semi_major, semi_minor, position_angle, **args):
      assert(semi_major >= semi_minor)  
      XY = (xc, yc)                  
      angle_mpl = position_angle     
      height = 2.0*semi_major        
      width  = 2.0*semi_minor        
      super().__init__(XY, width, height, angle=angle_mpl, **args)
    def Area(self, semi_major, semi_minor):
        area = pi * semi_major * semi_minor
        return area
    def Perimeter(self, semi_major, semi_minor):
        h = (semi_major-semi_minor)**2/(semi_major+semi_minor)**2
        P = pi*(semi_major+semi_minor)*(1+ 3*h/(10 + math.sqrt(4-3*h)))
        return P
    
```
