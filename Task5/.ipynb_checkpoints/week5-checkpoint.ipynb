Question 5.2

```python
import numpy as np

file = input("Enter a file you want to open: ")
data = np.loadtxt(file, comments = '#')

print(data)
#printing the text file I made

#taking collums of the matrix
X = data[:,0]; Y = data[:,1]

#making them arrays and taking their mean
print("Array X=", X)
print("Array Y=", Y)
meanX = np.mean(X)
meanY = np.mean(Y)
print(f"The mean of X is: {meanX}", f"The mean of Y is: {meanY}")

#getting the min and max of x
maxX = np.max(X)
minX = np.min(X)
print(f"The max of X is: {maxX} and the min of X is: {minX}")

#taking the min and max of Y
maxY = np.max(Y)
minY = np.min(Y)
print(f"The max of Y is: {maxY} and the min of Y is {minY}")

#making a new collum
Y2 = (Y-meanY)**2
print(Y2)
```

```python
import numpy as np

#Q1
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

c = np.concatenate((a,b))
print(c)
#the shape of this is a list

#Q2
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
d = np.concatenate((a, b))
print(d)
#the shape of this one is a matrix 4x2

e = np.concatenate((a, b), axis= 0)
print(e)
#The shape of this is a 4x2 matrix, made up of the two 2x2 matrices stacked vertically


f = np.concatenate((a, b), axis=1)
print(f)
#the shape of this is a 2x4 matrix with the two 2x2 matrices now stacked horizontally
```

```python
import numpy as np
from imageio import imwrite, imread
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure, show

m101 = imread("./m101BW.jpg")

# One can always write this image to your own device with imwrite()
imwrite('myimage.png', m101)  # From now on you can use myimage.png instead

fig = figure()
frame = fig.add_subplot(1,1,1)
m101 = imread('myimage.png')
print(m101.shape, m101.dtype)
frame.imshow(m101, cmap='jet')
frame.contour(m101, levels=[350, 400,500,550,600], colors=['purple' ,'y', 'b', 'r', 'g'])
show()

treshold = np.percentile(m101, 90)
filterdata = np.where(m101 >=treshold,m101,0)
plt.imshow(filterdata, cmap = 'jet')
show()
```

Question 5.10

```python
import numpy as np
from matplotlib import pyplot as plt

#Importin the file
file = "ngc6946"
data = np.loadtxt(file, comments = '#')

#making an array of the file and shaping it into the correct dimensions
filearray = np.array(data).reshape(40, 50, 70)

#takin the z-profile of the middle of the xy plane
S = filearray[:,25,35] 
print(f"A z-profile near the middle of the xy plane is: \n {S}")

#calculating different stats of the profile I extracted
maxZ = np.max(S)
print(f"the max of the z-profile is: {maxZ}")
minZ = np.min(S)
print(f"The min of the z-profile is: {minZ}")
sumZ= np.sum(S)
print(f"the sum of the z-profile is: {sumZ}")
meanZ = np.mean(S)
print(f"The mean of the z-profile is: {meanZ}")
stdevZ = np.std(S)
print(f"The standard deviaton of the z-profile is: {stdevZ}")

#making an arrange of the amout of different data points so I can make a plot
z = np.arange(len(S))

#plotting the 40 different values I have
plt.plot(z, S, label='Z-profile', color = 'purple')

#adding labels and text for clarity
plt.xlabel("Z plane")
plt.ylabel("Intensity")
plt.title("Z-profile near the middle of xy plane")

#puttin the calculated stats in the plot
stats = f"max = {maxZ:.2f}\nmin = {minZ:.2f}\nsum = {sumZ:.2f}\nmean = {meanZ:.2f}\nstd = {stdevZ:.2f}"
plt.text(1, maxZ*0.8, stats)
plt.legend()
plt.show()
```

Question 5.11

```python
import numpy as np
from matplotlib import pyplot as plt

#importing the file
file = "ngc6946"
data = np.loadtxt(file, comments = '#')

#making an array of the file and shaping into the correct dimensions and taking a singular plane
filearray = np.array(data).reshape(40, 50, 70)
plane = filearray[0, :, :]

#getting the mean and stdev so I can use those for the contours later
meanP = np.mean(plane)
stdevP = np.std(plane)
levels = [meanP,meanP + stdevP , meanP + 2*stdevP]

#plottin the image with the colorbar and the contours
plt.imshow(plane, origin = 'lower')
plt.colorbar()
plt.contour(plane, levels = levels , colors="white", linewidths = 1)
plt.show()
```

Question 5.12

```python
import sys
print(f"Python version: {sys.version}")

import numpy as np
print(f"NumPy version: {np.__version__}")

from scipy.constants import c

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      
    cms = c
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0) / self.caH
        return z
    def velocity(self):
        ckms = self.cms *1e-3
        v = self.redshift() * ckms
        return v


g1 = Galaxy("NGC1832", 3994.7)
print(f"The redshift of the galaxy is: {g1.redshift()}")
print(f"The correspondin velocity is: {g1.velocity()}")
```

```python
import numpy as np
from matplotlib.pyplot import figure, show
from matplotlib.patches import Ellipse

Nx = 400
Ny = 300
data = np.random.random((Ny,Nx))

fig = figure()
frame = fig.add_subplot(1,1,1, aspect=1)
xcenter = 200
ycenter = 150
width = 2*50; height= 2*100
angle = 30.0
# Note that the angle parameter is a keyword parameter
e1 = Ellipse((xcenter, ycenter), width, height, angle=angle, fill=False)
frame.add_patch(e1)
plt.imshow(data, interpolation='none', origin="lower", cmap='gray')
plt.show()
```
