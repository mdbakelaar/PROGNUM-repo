```python
from math import *

s = 0
n = [2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1, -2.0] #all the points given to us
n.remove(-2.0)                                       #removing the negative numbers because we don' t want to work with them 
for p in range(0, 8, 1):                             #in this for-loop we want to get the sum of all numbers
    s = s + n[p]

mean = s / 8                                         #we get the mean by dividing the sum of the number by the amount of points we have
print(f"The mean is: {mean}")

m = []                                               #make a new list for the points minus the mean
n = [2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1, -2.0]
n.remove(-2.0)                                       #in this for-loop we want to do the (points - mean) squared
for b in range(0, 8, 1):
    a = (n[b] - mean)**2
    m.append(a)

d = 0                                                #in this for-loop we want to sum all the (points - mean) squared so we can do our final calculation
for q in range(0, 8, 1):
    d = d + m[q]

std = sqrt(d/8)                                      #here we get the standard deviation by taking the previous sum and dividing it by 8
print(f"The standard deviation is: {std}")


import numpy as np

nplist = np.array(n)                                 #here we do the same calculations as before but now using numpy
npsum = np.sum(nplist)
npstd = np.std(nplist)
npmean = npsum / 8


print(f"The mean using numpy is: {npmean}")
print(f"The standard deviation using numpy is: {npstd}")
```

```python
def fibonacci_divisible_by_m(N, M):
    """
    Generates Fibonacci numbers less than N and returns those divisible by M.

    Args:
        N: The upper limit (exclusive) for the Fibonacci numbers.
        M: The divisor.

    Returns:
        A list of Fibonacci numbers less than N that are divisible by M.
    """
    if N <= 0 or M == 0:
        return []

    fib_numbers = []
    a, b = 0, 1
    
    # Generate Fibonacci numbers up to N
    while a < N:
        if a % M == 0:
            fib_numbers.append(a)
        a, b = b, a + b # Update values
        
    return fib_numbers

# Example usage:
limit = 100
divisor = 3
result = fibonacci_divisible_by_m(limit, divisor)
print(f"Fibonacci numbers less than {limit} divisible by {divisor}: {result}")

limit_2 = 200
divisor_2 = 5
result_2 = fibonacci_divisible_by_m(limit_2, divisor_2)
print(f"Fibonacci numbers less than {limit_2} divisible by {divisor_2}: {result_2}")
```

```python
from math import *
import numpy as np

distance = [18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
velocity = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]
x = distance
y = velocity

def pearson(X, Y):
    x = np.array(X)
    y = np.array(Y)
    rxy = (len(x)*sum(x*y)-sum(x)*sum(y))/(sqrt(len(x)*sum(x**2)-(sum(x))**2)*sqrt(len(x)*sum(y**2)-(sum(y)**2)))
    return rxy

print(pearson(x,y))
```
