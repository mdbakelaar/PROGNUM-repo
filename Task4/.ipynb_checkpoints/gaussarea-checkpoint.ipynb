```python
from matplotlib import pyplot as plt
import numpy as np
from scipy import integrate
from math import *

#Creating the fucntion 
def gauss(x, A, x0, sigma, z0):
    return A*np.exp(-(x-x0)**2/(2*sigma**2))+z0

#Allowing the user to input their own values
A = float(input("Enter the amplitude of the function: "))
x0 = float(input("Enter the position of the peak: "))
sigma = float(input("Enter the width of the peak: "))
z0 = float(input("Enter the offset in y: "))

#Making the range over which we want the points to be and putting the values in the function
rangee = np.linspace(-10, 10, 200)
func = gauss(rangee, A, x0, sigma, z0)

#Integrating over a chosen range
lowb = float(input("Enter the lower bound: "))
highb = float(input("Enter the high bound: "))
area = integrate.quad(lambda x: gauss(x, A, x0, sigma, z0), lowb, highb)
print(F"Quad derived area under the curve from your chosen boundaries is: {area}")

#Creating the range from 0 to 2.5 and putting it in the function to then shade this area in the plot
x_fill = np.linspace(lowb, highb, 100)
y_fill = gauss(x_fill, A, x0, sigma, z0)

#Plotting the function and shading the area of 0 to 2.5
plt.plot(rangee, func, alpha = 0.8, color = 'hotpink')
plt.fill_between(x_fill, y_fill, facecolor='pink', alpha=0.3, label=f'Integrated Area: {area}')
plt.xlim(-12, 12)
plt.legend(loc = 'upper left')
plt.show
```
