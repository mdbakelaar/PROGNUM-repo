### Monte Carlo Error Propagation (4 pt) <font color='red'><b>COURSEWORK</b></font>

One of the classical Physic Lab experiment is to use pendulum to calculate the local gravitational acceleration, ``g``, from the measurements of the pendulum length ``L`` and the oscillation period ``T``.

The formula is expressed as follows:

$$ g=\frac{4\pi^2 L}{T^2} $$

You have got a set of measurements of $L=1.00\pm0.01\ \mathrm{(m)}$ and $T=2.01\pm0.02\ \mathrm{(s)}$.

* Like in Physics Lab, use Tyler expansion described above to propagate the error. Write your derivation in Markdown.

* Calculate the error of the derived ``g``.

* Apply Monte Carlo error propagation to estimate the value and uncertainty of ``g``.

* Make appropriate comments.

* Compare the two results.

$$ \Delta g = \sqrt{(\frac{4\pi^2}{T^2} \Delta L)^2 + (\frac{-8\pi^2L}{T^3} \Delta T)^2} $$

````python
from math import *
import numpy as np

T = 2.01                                                                   #Our values for T and L and their errors
eT = 0.02
L = 1.00
eL = 0.01

G = (4*(pi**2)*L)/(T**2)                                                   #The formula for G
print(f"The calculated value for G using the given T and l: {G}")

errorG = sqrt((((4*(pi**2))/T**2)*eL)**2+(((-8*(pi**2)*L)/(T**3))*eT)**2)  #The formula to calculate the error in G
print(f"The error in G using the given error in T and L: {errorG}")

montcL = np.random.normal(1.00, 0.01, 100)                                 #Making normal distributions of our T and L 
arrayL = np.array(montcL)                                                  #Creating arrays of our T and L so we can randomly take numbers from them later to do the monte carlo errorprop.
montcT = np.random.normal(2.01, 0.02, 100)
arrayT = np.array(montcT)

indxL = np.random.randint(0, len(arrayL), 1)
indxT = np.random.randint(0, len(arrayT), 1)

listG = []
for i in range(0, 100, 1):                                                 #The monte carlo error propagation using randomly chosen points of our earlier created arrays of T and L
    indxL = np.random.randint(0, len(arrayL), 1)
    indxT = np.random.randint(0, len(arrayT), 1)
    G = (4*(pi**2)*arrayL[indxL])/(arrayT[indxT]**2)
    listG.append(G)

meanG = np.mean(listG)                                                     #The mean and standard deviation calculated using numpy and the values we got using monte carlo errorprop.
print(f"The calculated mean using monte carlo error propagation: {meanG}")
stdevG = np.std(listG)
print(f"The calculated standard deviation using monte carlo error propagation: {stdevG}")
````

### Area under a Gaussian (6 pt) <font color='red'><b>COURSEWORK</b></font>

A one dimensional Gaussian function is given by:

$$f(x) = A\ e^{\frac{-{(x-x_0)}^2}{2 \sigma^2}} + z_0$$

$A$ is the amplitude, $x_0$ is the position of the peak, $\sigma$ is the width of the peak
and $z_0$ is the offset in $y$.  

This function can be written in Python as follows:

```python
def gauss(x, A, x0, sigma, z0):
    return A*np.exp(-(x-x0)**2/(2*sigma**2))+z0
```

1. Set in the calling environment:
```python
    A = 1.0
    x0 = 0.0
    sig = 2.0
    z0 = 0.0
```
2. Plot this function between -10 and 10 with 200 samples.
3. Use function `scipy.integrate.quad()` to find the area between ``x=0`` and ``x=2.5``.
4. Use function `matplotlib.pyplot.fill_between()` to mark that area in your plot.
5. Calculate the area between $-\infty$ and $\infty$ ($\approx 5.01$).
6. Given is the integral for the total area under the curve:

    $$\int _{-\infty }^{\infty }Ae^{-(x-x_0)^{2}/(2\sigma^{2})}\,\mathrm{d}x=A \sigma \cdot {\sqrt {2\pi }}$$

   Calculate this area for the values of the parameters given above and prove with ``quad()`` that the two are the same.

7. Make the last part an executable script, called ``gaussarea``. The user can enter values for ``A``, ``x0``, ``sig`` and ``z0`` and the integration limits. The programme then calculates the area under the curve between the integration limits using function ``quad()``. The programme must includes a plot of the curve, the integration area, and the value of that area in a legend.
8. Insert the generated plot in a separate markdown cell below your code cell, and upload your script to GitHub. Provide your GitHub website link in the notebook too for TA to check. 
   
**Answers**:   

````python
#The code for this question using only values I decided and put in
from matplotlib import pyplot as plt
import numpy as np
from scipy import integrate
from math import *

#Creating the fucntion 
def gauss(x, A, x0, sigma, z0):
    return A*np.exp(-(x-x0)**2/(2*sigma**2))+z0

#The chosen values by me
A = 1.0
x0 = 0.0
sigma = 2.0
z0 = 0.0

#Making the range over which we want the points to be and putting the values in the function
rangee = np.linspace(-10, 10, 200)
func = gauss(rangee, A, x0, sigma, z0)

#Integrating over x=0 to x=2.5
area = integrate.quad(lambda x: gauss(x, A, x0, sigma, z0), 0.0, 2.5)
print(F"Quad derived area under the curve from x=0 to x=2.5 is: {area}")

#Creating the range from 0 to 2.5 and putting it in the function to then shade this area in the plot
x_fill = np.linspace(0, 2.5, 100)
y_fill = gauss(x_fill, A, x0, sigma, z0)


#Calculating the integral from negative infinity to positive infinity
xlo = -np.inf
xhi = np.inf
infarea = integrate.quad(lambda x: gauss(x, A, x0, sigma, z0), xlo, xhi)
print(f"Quad derived area under the curve from neg infinity to pos infinity is: {infarea}")


#The integral is equal to the following function, let's check if they give the same answer given the same input.
nonintgfunc = A * sigma * sqrt(2*pi)
print(f"The area under the function given by a function and not quad is: {nonintgfunc: .2f}")
print("You can see that both the quad and the non quad function give the same answer")

#Plotting the function and shading the area of 0 to 2.5
plt.plot(rangee, func, alpha = 0.8, color = 'darkgreen')
plt.fill_between(x_fill, y_fill, facecolor='g', alpha=0.3, label=f'Integrated Area: {area}')
plt.xlim(-12, 12)
plt.legend(loc = 'upper left')
plt.show
````

### Calculate integral of a user supplied function (6 pt) <font color='red'><b>COURSEWORK</b></font>

Write a programme that calculates the integral of any function from the user input using ``eval()``, which is a string representing a function and a variable (floating number) for which the function will be evaluated.

* Call from ``numpy import sin, cos, exp, pi`` to apply these functions and constants in your code.
* Use ``scipy.integrate.quad()`` to calculate the integral of the formula below between $x=0$ and $x=\pi$
* Use Monte Carlo Integration to calculate the integral of the formula below for $x=0$ and $x=\pi$

$$ f(x)=x^4+e^{\sin{x}+\cos{x}} $$

Make the function robust with try/except clauses. Demonstrate exceptions for wrong expressions and unknown functions in an expression (NameError) etc. Print messages that tell a user what they have done wrong!

Make the above code an executable script called ``interate_user_function``, and upload to GitHub.

* Create a separate cell, and use SymPy expression to calculate the integral of the formula below for $x=0$ and $x=\pi$.

````python
import numpy as np
from scipy.integrate import quad
from numpy import sin, cos, exp, pi

# letting you enter a function
userfunc = input("Enter function in terms of x (e.g., x**4 + exp(sin(x) + cos(x))): ")

# evaluate what the user uses as a function
def f(x):
    return eval(userfunc, {"x": x, "sin": sin, "cos": cos, "exp": exp, "pi": pi})

try:
    # using quad function to integrate
    area = quad(f, 0, pi)
    print(f"Integral from 0 to pi using quad gives: {area}")

    # Monte Carlo parameters
    N = 100000
    a = 0
    b = pi

    # random numbers
    g = np.random.uniform(a, b, N)
    fx = f(g) 

    # Monte Carlo integration
    mcintegral = (b - a)/len(g) * np.sum(fx)

    print(f"Integral from 0 to pi using Monte Carlo gives: {mcintegral}")
    
#If the user inputs a wrong function and would normally get an error there are now except functions to tell the user what they did wrong
except NameError as e:
    print("Error: Unknown function or variable used.")
    print(e)

except SyntaxError:
    print("Error: Invalid mathematical expression.")

except Exception as e:
    print("Something went wrong:")
    print(e)
````
