```python
from math import *

a = input("enter any integer for a: ")
b = input("enter any integer for b: ")
c = input("enter any integer for c: ")

a = int(a)
b = int(b)
c = int(c)

D = b**2 - 4*a*c

if D>0:
    x1 = (-b + sqrt(D))/(2*a)
    x2 = (-b - sqrt(D))/(2*a)
    e = f"the answers are {x1} and {x2}"
    print(e)

elif D==0:
    x1 = (-b + sqrt(D))/(2*a)
    g = f"the answer is {x1}"
    print(g)

else:
    print("There is no solution")
```
