# Energy and frequency

With Planck's constant $h$, we write the relation between energy and frequency as:

$$E = hf$$

$$\alpha = \lambda \Delta x$$

$$E = hc/\lambda$$

$$F = G \frac{m_{1}m_{2}}{r^2}$$

$$\sin{2x} = 2 \sin{x}\cos{x} = \frac{2\tan{x}}{1 + \tan^2{x}}$$

```python
from math import *

alpha_d = 30                  # degrees
alpha_r = radians(alpha_d)    # Now in radians

ans1=sin(2*alpha_r)
ans2=2*sin(alpha_r)*cos(alpha_r)
ans3=(2*tan(alpha_r))/(1 + tan(alpha_r)**2)

print (ans1,ans2,ans3)
```
