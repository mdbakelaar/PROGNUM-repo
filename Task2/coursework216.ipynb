```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
set1 = set(night1)
list1 = list(set1)
print (f"set1: {set1}", f"list1: {list1}")

night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
set2 = set(night2)
list2 = list(set2)
print (f"set2: {set2}", f"list2: {list2}")


inters = set.intersection(set1, set2)
print(f"Intersection set1, set2: {inters}")
interset = list(inters)
print (interset)

diff = set1.difference(set2)
print(f"Difference between set1,set2: {diff}")
list3 = list(diff)
print (f"difference between night1 and night2: {list3}")

set4 = set1.union(set2)
print (set4)
list4 = list(set4)
print (f"list of all values in both night1 and night2: {list4}")

night1 = set.intersection(set1, set2)
list6 = list(night1)
print (f"night1 without the values not in night2: {list6}")
```

```python
planetsmass = { 'Mercury' : 0.33,
                'Venus'   : 4.87, 'Earth'   : 5.98,
                'Mars'    : 0.65, 'Jupiter' : 1900,
                'Saturn'  :  570, 'Uranus'  :  87,
                'Neptune' :  100, 'Pluto'   :  0.7 }

masses = planetsmass.values()
planets = planetsmass.keys()
planetlist = list(planets)
print (f"list of planets: {planetlist}")
maslist = list(masses)
print(f"list of masses; {maslist}")

venus = maslist[1]
earth = maslist[2]


vine = venus / earth
print (f"Mass of venus in earthmasses: {vine}")

planetlist.remove('Pluto')
maslist.remove(0.7)
print (f"Planets without pluto: {planetlist}", f"masses without pluto's: {maslist}")

planetlist.append('planet9')
maslist.append(29.3)
print (f" planets with extra planet 9: {planetlist}", f"masslist with mass of planet 9: {maslist}")

del planetsmass['Pluto']
planetsmass ['planet 9'] = 29.3
print (planetsmass)
```
